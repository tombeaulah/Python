```python
# IMPORT REQUIRED PACKAGES 

import time
import osl
from dotenv import load_dotenv
from google import genai
```

```python
# LOAD KEYS & FILES

#load_dotenv(dotenv_path=r"C:\Users\tombe\Python\GEMINI_API_KEY.env")
load_dotenv(dotenv_path="../GEMINI_API_KEY.env")
client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

#os.environ["GEMINI_API_KEY"] = "YOUR_ACTUAL_KEY_HERE"
#client = genai.Client(api_key=os.environ["GEMINI_API_KEY"])

video_to_assess = r"C:\Users\tombe\Python\Data\BLINK\26-03\26-03-27\20-52-11_FluffballGNT2XP0141632NKX_253.mp4"
fluffball_path = r"C:\Users\tombe\Python\Data\BLINK\20260328-Fluffball_Image.jpg"


# Upload the video
print("Uploading video...")
video_file = client.files.upload(file=video_to_assess)  # <-- change this path
```

```python
# CALL GEMINI TO REVIEW VIDEO FILE

# Wait for processing (same pattern as your generate_content, just async)
print("Processing...")
while video_file.state.name == "PROCESSING":
    time.sleep(3)
    video_file = client.files.get(name=video_file.name)
    fluffball_photo = client.files.upload(file=fluffball_path)  # your cutout photo

str_instructions = """
Did a bird visit the fluffball in this video?
The fluffball is shown in the still image provided.  It is a cage-like structure filled with cotton-like material for birds to pull nesting material from.
Please tell time of bird first coming into the video; when the bird is proximate to the fluffball, how long the bird visits the fluffball for, and at what time it exits the proximity of the fluffball
Identify any noteworthy behaviour please
Tell me if no bird visit!
I'm only interested if a bird actually gets close to the fluffball
Identify the bird species and gender if possible
"""


# Your existing pattern, just with video added to contents
response = client.models.generate_content(
    model="models/gemini-2.5-flash",
    contents=[
        str_instructions,
        fluffball_photo,
        video_file
    ]
)
print(response.text)

# Cleanup
client.files.delete(name=video_file.name)
```
