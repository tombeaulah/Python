```python

fruits = ["apple", "banana", "orange"]


def shopping_list(item):

    if item in fruits:

        def fruits_price(q, p= 1.5):

            return q * p
            
        
        return fruits_price

    else:

         print("Unknown item.")



pricing = shopping_list("apple")
```
