```python
import ffmpeg
import os

input_folder = "videos"    # folder with your .avi files
output_folder = "converted"  # where MP4s will be saved

os.makedirs(output_folder, exist_ok=True)

avi_files = [f for f in os.listdir(input_folder) if f.endswith(".avi")]
print(f"Found {len(avi_files)} AVI files to convert...")

for filename in avi_files:
    input_path = os.path.join(input_folder, filename)
    output_filename = filename.replace(".avi", ".mp4")
    output_path = os.path.join(output_folder, output_filename)

    print(f"Converting {filename}...")
    
    try:
        ffmpeg.input(input_path).output(
            output_path,
            vcodec="libx264",
            crf=28,
            acodec="aac"
        ).run(overwrite_output=True)
        print(f"  ✓ Done → {output_filename}")
    
    except Exception as e:
        print(f"  ✗ Failed: {filename} — {e}")

print("All done!")
```
