```python
import re

the_str = "5t8u4e9y3q7h 789"

regex1 = r"(\d\w)+\s\d+"
regex2 = r"(?:\d\w)+\s\d+"

matches1 = re.findall(regex1, the_str)
matches2 = re.findall(regex2, the_str)
matches3 = re.search(regex1, the_str)

print(matches1)
print(matches2) 
print(matches3)
```

```python
import re

the_str = "The amount of time I took walking the dogs yesterday was 178 and 23 minutes"

regex1 = r"(\d)+"
regex2 = r"(?:\d)+"

matches1 = re.findall(regex1, the_str)
matches2 = re.findall(regex2, the_str)
matches3 = re.search(regex1, the_str)

print(matches1)
print(matches2) 
print(matches3)
```

```python
import re

my_str = "There are 6 Anna's, 2 Allen's and no Costa's"

regex1 = r"\d+\sAnna's|Allen's|Costa's"
regex2 = r"\d+\s(Anna's|Allen's|Costa's)"
regex3 = r"(\d+)\s(Anna's|Allen's|Costa's)"

matches1 = re.findall(regex1, my_str)
matches2 = re.findall(regex2, my_str)
matches3 = re.findall(regex3, my_str)
  
print("This shows the output when regex is an OR pattern, but the \\d+ ends up belonging only to the Anna's: \n" + str(matches1) + "\n")
print("This shows the output when regex uses a capturing group for the \\d+: \n" + str(matches2)+ "\n")
print("This shows the output when regex uses capturing groups for both the \\d+ and the names: \n" + str(matches3)+ "\n")
```

```python
import re

my_str = "Two codes: 02-08-20-34-0897 and 55-32-03-04-1798"

regex1 = r"(?:\d{2}-){4}(\d{4})"

matches1 = re.findall(regex1, my_str)

print("This shows the output using a non-capturing group: \n" + str(matches1) + "\n")
```

```python
my_str = "I moved on 01-12-2025."

date_moved = re.search(r'(\d{1,2})-(\d{2})-(\d{4})', my_str)
print(date_moved.group(3))
print(date_moved.group(0))
```

```python
my_str = "I moved on 01-12-2025."

date_moved = re.search(r'(?P<day>\d{1,2})-(?P<month>\d{2})-(?P<year>\d{4})', my_str)
print(date_moved.group("year"))
print(date_moved.group("month"))
print(date_moved.group("day"))
```
