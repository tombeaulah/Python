```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error as MSE, roc_auc_score
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
auto_mpg = fetch_ucirepo(id=9) 
  
# data (as pandas dataframes) 
X = auto_mpg.data.features 
y = auto_mpg.data.targets 

X = X.dropna()
y=y.loc[X.index]

print(X.info())
print(y.info())

# metadata 
print(auto_mpg.metadata) 
  
# variable information 
print(auto_mpg.variables) 
```

```python
SEED=1

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Instantiate rf
gbt = GradientBoostingRegressor(n_estimators=300, max_depth=1, random_state=2026)

# Fit rf to the training set    
gbt.fit(X_train, y_train) 

# Predict the test set labels
y_pred = gbt.predict(X_test)

# Evaluate the test set RMSE
gbt_mse_test = MSE(y_test, y_pred)

gbt_rmse_test = gbt_mse_test ** 0.5
 
# Print rmse_test
print('Test set RMSE: {:.2f}'.format(gbt_rmse_test))
```
