```python
from sklearn.feature_extraction.text import TfidfVectorizer
import scipy.sparse

# Your toy corpus — a list of documents (strings)
corpus = [
    "the cat sat on the mat",
    "the dog sat on the log",
    "cats and dogs are great pets",
    "the mat is on the floor",
]

# Create the vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform — returns a sparse CSR matrix
X = vectorizer.fit_transform(corpus)

print(type(X))          # <class 'scipy.sparse.csr_matrix'>
print(X.shape)          # (4 documents, N unique tokens)
print(X)                # Shows (row, col) : tfidf_value for non-zero entries

# To see the vocabulary (token → column index mapping)
print("the vocab is: " + str(vectorizer.vocabulary_))

# To see feature names in column order
print(vectorizer.get_feature_names_out())

words = vectorizer.get_feature_names_out()

toy_data = X
```

```python
# Import NMF
from sklearn.decomposition import NMF

# Create an NMF instance: model
model = NMF(n_components=3)

# Fit the model to articles
model.fit(toy_data)

print(model.components_.round(2))
print("model components shape is: " + str(model.components_.shape))

# Transform the articles: nmf_features
nmf_features = model.transform(toy_data)

# Print the NMF features
print(nmf_features.round(2))
print("model features shape is: " + str(nmf_features.shape))

print(toy_data)

```

```python
# Import pandas
import pandas as pd

# Create a DataFrame: components_df
components_df = pd.DataFrame(model.components_,columns=words)

# Print the shape of the DataFrame
print(components_df.shape)

n_row = 2 #specify which component to select for display

# Select row 3: component
component = components_df.iloc[n_row,:]
print(f"extract and display of component {n_row}: \n" + str(component) + "\n")
# Print result of nlargest
print("highest ranked words: \n" + str(component.nlargest(4)))
```
