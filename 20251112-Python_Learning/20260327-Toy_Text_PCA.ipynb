```python

# Import Necessary Packages
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline
import pandas as pd
```

```python
# CONSTRUCT THE DATA
str1 = "Hummingbirds were observed to start pulling material on 27th March 2026"
str2 = "Tom and Rosemary started the hummingbird project in or around March 2026"
str3 = "Tom and Rosemary were delighted to observe that the Hummingbirds were keen to use the new fountain feature"
str4 = "The fountain was an iterative sub-project"
str5 = "The Hummingbirds seem to tolerate each other more during the early morning and late evening, and can be seen to line up on the feeders, either two or even three together.  Very occassionally 4 have been seen together too!!"

text_array = [str1,str2,str3,str4,str5]

print (text_array)

int_sentence = range(1,6)
```

```python

# THIS CELL PRODUCES A SPARSE MATRIX OF WORDS USED

# Create a TfidfVectorizer: tfidf
tfidf = TfidfVectorizer() 

# Apply fit_transform to document: csr_mat
csr_mat = tfidf.fit_transform(text_array)

# Print result of toarray() method
print(csr_mat.toarray())

# Get the words: words
words = tfidf.get_feature_names_out()

# Print words
print(words)
```

```python
# Create a TruncatedSVD instance: svd
svd = TruncatedSVD(n_components=20)

# Create a KMeans instance: kmeans
kmeans = KMeans(n_clusters=3)

# Create a pipeline: pipeline
pipeline = make_pipeline(svd,kmeans)
```

```python

# Fit the pipeline
pipeline.fit(csr_mat)

# Calculate the cluster labels: labels
labels = pipeline.predict(csr_mat)

# Create a DataFrame aligning labels and titles: df
df = pd.DataFrame({'label': labels, 'Sentence': int_sentence})

# Display df sorted by cluster label
print(df.sort_values(by='label'))
```
