```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

```python
df_data = pd.read_excel(r"C:\Users\tombe\Python\Data\US_Crime\20260323-NIBRS_Table_27_Crimes_Against_Persons_Offenses_Offense_Category_by_State_2024.xlsx",0)
df_data = df_data.drop(index=0)

df_X = df_data["Percent_total_offences"]
arr_X = df_X.values.reshape(-1,1)
df_y = df_data["State"]
```

```python
print (arr_X)
print (arr_X.shape)
print (df_X.shape)
print (df_X.info())
```

```python
mergings = linkage(arr_X, method='complete')
#print(mergings)
```

```python

plt.figure(figsize=(12,6))
plt.xlabel('States',fontsize=16)

dendrogram(mergings,
labels=df_y.values,
leaf_rotation=90,
leaf_font_size=10)

#plt.ylim(0,0.0001) # optional quick way to zoom using the y-scale

plt.show()
```

```python
#extract labels usng the fcluster method from the mergings matrix
labels = fcluster (mergings, 0.0025, criterion='distance')
#print(labels) # optional print of the cluster labels assigned to each state

pairs = pd. DataFrame({'labels': labels, 'states': df_y})
#print(pairs.sort_values('states')) # optional print of the clusters that the states are assigned to at the given cutoff value
```
