```python
labels = ['Country', 'City' , 'Currency' ]
sublist = [['England', 'London', 'GBP' ], ['France', 'Paris', 'EUR' ], ['Germany', 'Berlin', 'EUR' ]]

def list_to_dict(x, y):
lists = zip(x, y)
new_dict = dict(lists)

return new_dict

list_of_dicts = [list_to_dict(labels, items) for items in sublist]

print (list_of_dicts)
```

```python
import re

the_str = "5t8u4e9y3q7h 789"

regex1 = r"(\d\w)+\s\d+"
regex2 = r"(?:\d\w)+\s\d+"

matches1 = re.findall(regex1, the_str)
matches2 = re.findall(regex2, the_str)
matches3 = re.search(regex1, the_str)

print(matches1)
print(matches2) 
print(matches3)

print ( matches1.count() )

```
