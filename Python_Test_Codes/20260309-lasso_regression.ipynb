```python

from sklearn.linear_model import Lasso
from sklearn.datasets import load_diabetes
import pandas as pd
import matplotlib.pyplot as plt


diabetes_df = load_diabetes(as_frame=True)
test_list = diabetes_df.feature_names
diabetes_df = diabetes_df.frame

col_names = {'s1':'tc', 's2':'ldl','s3':'hdl', 's4':'tch','s5':'ltg','s6':'glu'}

diabetes_df = diabetes_df.rename(columns = col_names)

print(diabetes_df)


```

```python
X = diabetes_df.drop(["target"], axis=1).values
y = diabetes_df["target"].values
y = y.reshape(-1,1)

names = diabetes_df.drop(["target"], axis=1).columns
lasso = Lasso(alpha=0.1)
lasso_coef = lasso.fit(X, y).coef_


print (y.shape)

print(names)
print(lasso_coef)
```

```python
plt.bar(names, lasso_coef)
plt.xticks(rotation=45)
plt.show()
```
