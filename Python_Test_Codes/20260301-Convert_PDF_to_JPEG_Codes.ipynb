```python
from pdf2image import convert_from_path
from PIL import Image

# -----------------------------
# CONFIG
pdf_path = "C:\\Users\\tombe\\Google Drive\\20241126-Liver_Function_Test_Result_November_2024.pdf"  # your PDF file
dpi = 300
max_dimension = 4000  # max width or height in pixels
output_prefix = "combined"
# -----------------------------

# Convert PDF pages to images
pages = convert_from_path(pdf_path, dpi=dpi)

combined_images = []

# Combine pages 2 at a time, side by side
for i in range(0, len(pages), 2):
    page1 = pages[i].convert("RGB")
    page2 = pages[i+1].convert("RGB") if i+1 < len(pages) else None

    if page2:
        # Combine horizontally
        total_width = page1.width + page2.width
        max_height = max(page1.height, page2.height)
        combined = Image.new('RGB', (total_width, max_height), color=(255,255,255))
        combined.paste(page1, (0, 0))
        combined.paste(page2, (page1.width, 0))
    else:
        combined = page1

    # Resize if larger than max_dimension
    w, h = combined.size
    scale = min(max_dimension / w, max_dimension / h, 1.0)  # never upscale
    if scale < 1.0:
        new_size = (int(w*scale), int(h*scale))
        combined = combined.resize(new_size, Image.LANCZOS)

    combined_images.append(combined)

# Save all combined images as website-friendly JPEGs
for idx, img in enumerate(combined_images, 1):
    img.save(f"{output_prefix}_{idx}.jpg", "JPEG", quality=95, progressive=False, optimize=True)

print(f"Done! Created {len(combined_images)} JPG file(s).")
```
