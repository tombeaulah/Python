```python
import fitz  # PyMuPDF


# define function
def add_text_box(input_pdf, page_number, text, output_pdf):
    doc = fitz.open(input_pdf)
    page = doc[page_number - 1]

     #define start point
    x0_y0 = [12, 4]
    x1_y1 = [20, 8]

    coords = x0_y0 + x1_y1

    cm_pts = 28.35  # cm to points

    coords = [coord * cm_pts for coord in coords]

    # Define a rectangle for the box (adjust coordinates as needed)
    rect = fitz.Rect(coords)  # x0,y0,x1,y1
    # 

    # Draw the rectangle (border)
    page.draw_rect(rect, color=(1, 0, 0), width=1)  # red border

    # Insert text inside the rectangle
    page.insert_textbox(
        rect,
        text,
        fontsize=18,
        fontname="helv",
        color=(1, 0, 0),
        align=0  # left align
    )

    # Save the PDF
    doc.save(output_pdf)
    doc.close()
```

```python
str_text = "I rang Avanti on 1st April 2026 to change the address to HU17 9SS"

working_pdf = r"C:\Users\tombe\Google Drive\Administration\Insurance\Travel Insurance\20260212-Avanti_Mailshot.pdf"
new_pdf = r"C:\Users\tombe\Google Drive\Administration\Insurance\Travel Insurance\20260212-Avanti_Mailshot1.pdf"

add_text_box(working_pdf, 1, str_text, new_pdf)
```
